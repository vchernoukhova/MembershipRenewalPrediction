```python
# NEW VARIABLES: ANNIE
## previous record variables

memberships['LastMemCategory'] = memberships.sort_values(by = 'transDateTime').groupby('memId').memCategory.shift(1)
memberships['LastChannel'] = memberships.sort_values(by = 'transDateTime').groupby('memId').channel.shift(1)

 

## variables for fields with number of days

memberships['TransEndDate'] = memberships.sort_values(by = 'transDateTime').groupby('memId').transDateTime.shift(-1)
memberships['TransEndDate'] = memberships['TransEndDate'].fillna(np.datetime64(datetime.now()))

memberships['NumActiveDays'] = (memberships['TransEndDate'] - memberships['transDateTime']).dt.days
memberships.loc[memberships.transType == 'Dropped',['NumActiveDays']] = 0

memberships['CustomerTenure'] = memberships.sort_values(by = 'transDateTime').groupby('memId').NumActiveDays.cumsum()
memberships['FirstJoinDate'] = memberships.sort_values(by = 'transDateTime').groupby('memId').transDateTime.transform('min')

memberships['DaysSinceJoined'] = (memberships['transDateTime'] - memberships['FirstJoinDate']).dt.days

 

## counting transTypes and giftTypes

memberships.loc[memberships['transType'] == 'Renewal', 'Renewal_Ind'] = 1
memberships['Renewal_Ind'] = memberships['Renewal_Ind'].fillna(0)
memberships['NumberOfRenewals'] = memberships.sort_values(by = 'transDateTime').groupby('memId').Renewal_Ind.cumsum()

 

memberships.loc[memberships['transType'] == 'Rejoined', 'Rejoined_Ind'] = 1
memberships['Rejoined_Ind'] = memberships['Rejoined_Ind'].fillna(0)
memberships['NumberOfRejoins'] = memberships.sort_values(by = 'transDateTime').groupby('memId').Rejoined_Ind.cumsum()

 

memberships.loc[memberships['transType'] == 'Dropped', 'Dropped_Ind'] = 1
memberships['Dropped_Ind'] = memberships['Dropped_Ind'].fillna(0)
memberships['NumberOfDrops'] = memberships.sort_values(by = 'transDateTime').groupby('memId').Dropped_Ind.cumsum()
 

memberships.loc[memberships['transType'] == 'Upgrade', 'Upgrade_Ind'] = 1
memberships['Upgrade_Ind'] = memberships['Upgrade_Ind'].fillna(0)
memberships['NumberOfUpgrades'] = memberships.sort_values(by = 'transDateTime').groupby('memId').Upgrade_Ind.cumsum()

 
memberships.loc[memberships['transType'] == 'Downgrade', 'Downgrade_Ind'] = 1
memberships['Downgrade_Ind'] = memberships['Downgrade_Ind'].fillna(0)
memberships['NumberOfDowngrades'] = memberships.sort_values(by = 'transDateTime').groupby('memId').Downgrade_Ind.cumsum()


memberships.loc[memberships['giftType'] == 'Cash', 'Cash_Ind'] = 1
memberships['Cash_Ind'] = memberships['Cash_Ind'].fillna(0)
memberships['NumberOfCash'] = memberships.sort_values(by = 'transDateTime').groupby('memId').Cash_Ind.cumsum()


memberships.loc[memberships['giftType'] == 'Pledge', 'Pledge_Ind'] = 1
memberships['Pledge_Ind'] = memberships['Pledge_Ind'].fillna(0)
memberships['NumberOfPledges'] = memberships.sort_values(by = 'transDateTime').groupby('memId').Pledge_Ind.cumsum()

 

## sum gift amounts

memberships['NumericGiftAmount'] = memberships['giftAmount'].replace('[\$,]', '', regex=True).astype(float)
memberships['NumericGiftAmount'] = memberships['NumericGiftAmount'].fillna(0)
memberships['CumulativeGiftAmount'] = memberships.sort_values(by = 'transDateTime').groupby('memId').NumericGiftAmount.cumsum()
 

## counting channels

memberships.loc[memberships['channel'] == 'Front.of.House', 'FoH_Ind'] = 1
memberships['FoH_Ind'] = memberships['FoH_Ind'].fillna(0)
memberships['NumberOfFoH'] = memberships.sort_values(by = 'transDateTime').groupby('memId').FoH_Ind.cumsum()
 

memberships.loc[memberships['channel'] == 'Telemarketing', 'Telemarketing_Ind'] = 1
memberships['Telemarketing_Ind'] = memberships['Telemarketing_Ind'].fillna(0)
memberships['NumberOfTelemarketing'] = memberships.sort_values(by = 'transDateTime').groupby('memId').Telemarketing_Ind.cumsum()


memberships.loc[memberships['channel'] == 'Web', 'Web_Ind'] = 1
memberships['Web_Ind'] = memberships['Web_Ind'].fillna(0)
memberships['NumberOfWeb'] = memberships.sort_values(by = 'transDateTime').groupby('memId').Web_Ind.cumsum()
 

memberships.loc[memberships['channel'] == 'Direct.Mail', 'DM_Ind'] = 1
memberships['DM_Ind'] = memberships['DM_Ind'].fillna(0)
memberships['NumberOfDM'] = memberships.sort_values(by = 'transDateTime').groupby('memId').DM_Ind.cumsum()


#Next Action
memberships['NextAction'] = memberships.sort_values(by = 'transDateTime').groupby('memId').transType.shift(-1)
memberships.loc[memberships['NextAction'] == 'Renewal', 'IsRenewed'] = 1
memberships['IsRenewed'] = memberships['IsRenewed'].fillna(0)
```

```python
# NEW VARIABLES: VICKIE

memberships['RecordNum'] = memberships[['transDateTime','memId']].sort_values(by = 'transDateTime').groupby('memId').cumcount()+1
memberships['TotRecordNum'] = memberships.groupby('memId')['memId'].transform('size')

memberships.loc[memberships['MemAtrCat_1_Description'].isna() == True, 'Promo'] = 0
memberships.loc[memberships['MemAtrCat_1_Description'].isna() == False, 'Promo'] = 1

cat_split = pd.DataFrame(memberships['memCategory'].str.split(' - ',1).tolist(),columns = ['memType','memLength'])
memberships['memType'] = cat_split['memType']
memberships['memLength'] = cat_split['memLength']

memberships.loc[memberships['memLength'] == '1 Year', 'Membership_end_date' ] = memberships['transDateTime'] + timedelta(days=365)
memberships.loc[memberships['memLength'] == '2 Years', 'Membership_end_date' ] = memberships['transDateTime'] + timedelta(days=2*365)
memberships['DaysBeforeExpiration'] = (memberships['Membership_end_date'] - np.datetime64(datetime.now())).dt.days

memberships['postalCode']=memberships['postalCode'].astype('str')
postal_split = pd.DataFrame(memberships['postalCode'].str.split(' ',1).tolist(), columns = ['postal_first','postal_second'])
memberships['postal_first'] = postal_split['postal_first']

memberships.loc[memberships['NextAction'] == 'Dropped', 'NotDropped'] = 0
memberships.loc[memberships['NextAction'] != 'Dropped', 'NotDropped'] = 1
```

```python
import pandas as pd
import numpy as np
import scipy.stats as st

from boostaroota import BoostARoota

from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_curve

from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC 


np.random.seed(0)
```

```python
trainTreated = pd.read_csv('dTrainTreated.csv')
trainTreated = trainTreated.drop('Unnamed: 0', 1)
trainTreated.set_index('memId', inplace=True)
```

```python
list(trainTreated)
```

```python
# BoostARoota from: https://github.com/chasedehan/BoostARoota
br = BoostARoota(metric='logloss')
train = trainTreated.loc[:, trainTreated.columns != 'NotDropped']
label = trainTreated["NotDropped"]

br.fit(train,label)
len(train.columns)
len(br.keep_vars_)
new_train = br.transform(train)

print("Original training set has " + str(train.shape) + " dimensions. \n" +\
"BoostARoota with .fit() and .transform() reduces to " + str(new_train.shape) + " dimensions.")
```

```python
# distributions of params from: http://danielhnyk.cz/how-to-use-xgboost-in-python/

XGBmodel = XGBClassifier()
# XGBmodel.fit(new_train, label)

one_to_left = st.beta(10, 1)  
from_zero_positive = st.expon(0, 50)
XGB_search_params = {  
    "n_estimators": st.randint(3, 40),
    "max_depth": st.randint(3, 40),
    "learning_rate": st.uniform(0.05, 0.4),
    "colsample_bytree": one_to_left,
    "subsample": one_to_left,
    "gamma": st.uniform(0, 10),
    'reg_alpha': from_zero_positive,
    "min_child_weight": from_zero_positive,
}


XGBsearch = RandomizedSearchCV(XGBmodel, XGB_search_params, cv=3, scoring = 'roc_auc', n_iter=20)
XGBsearch.fit(new_train, label)
```

```python
testTreated = pd.read_csv('dTestTreated.csv')
testTreated = testTreated.drop('Unnamed: 0', 1)
testTreated.set_index('memId', inplace=True)
```

```python
# Get the Boruta reduced column list
boruta_columns = list(new_train)
# Filter test set columns and remove the columns excluded by Boruta
test_br = testTreated[np.intersect1d(testTreated.columns, boruta_columns)]
# Ensure the order of the test set columns is aligned with the training set of the model
test_br = test_br[boruta_columns]
```

```python
XGBparams = XGBsearch.best_params_
XGBmodel.set_params(**XGBparams)
XGBmodel.fit(new_train, label)

y_pred_test = XGBmodel.predict(test_br)
test_br['NotDropped'] = y_pred_test
```

```python
from sklearn.metrics import accuracy_score
acc = 100*accuracy_score(y_test, y_pred_test)
acc
```

```python
#ROC + AUC
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred_test)*100
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_test)
```
